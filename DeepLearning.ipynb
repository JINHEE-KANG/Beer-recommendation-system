```python
user_enc=LabelEncoder()
ratings['user']=user_enc.fit_transform(ratings['userid'].values)
n_users=ratings['user'].nunique()
```

```python
item_enc=LabelEncoder()
ratings['beer']=user_enc.fit_transform(ratings['beerid'].values)
n_beers=ratings['beer'].nunique()
```

```python
X=ratings[['user','beer']].values
y=ratings['rating'].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
n_factors=50

X_train_array=[X_train[:,0],X_train[:,1]]
X_test_array=[X_test[:,0],X_test[:,1]]
```

```python
from keras.models import Model
from keras.layers import Input,Reshape,Dot
from keras.layers.embeddings import Embedding
from keras.optimizers import Adam
from keras.regularizers import l2

def RecommenderV1(n_users,n_movies,n_factors):
    user=Input(shape=(1,))
    u=Embedding(n_users,n_factors,embeddings_initializer='he_normal',
               embeddings_regularizer=l2(1e-6))(user)
    u=Reshape((n_factors,))(u)
    
    beer=Input(shape=(1,))
    m=Embedding(n_beers,n_factors,embeddings_initializer='he_normal',
               embeddings_regularizer=l2(1e-6))(beer)
    m=Reshape((n_factors,))(m)
    
    x=Dot(axes=1)([u,m])
    
    model=Model(inputs=[user,beer],outputs=x)
    opt=Adam(lr=0.001)
    model.compile(loss='mean_squared_error',optimizer=opt)
    
    return model
```

```python
model=RecommenderV1(n_users,n_beers,n_factors)
model.summary()
```

```python
history=model.fit(x=X_train_array,y=y_train,batch_size=32,epochs=5,verbose=1,
                 validation_data=(X_test_array,y_test))
```

```python
model.evaluate(X_test_array,y_test,batch_size=32)
```

```python
from keras.layers import Add,Activation,Lambda

class EmbeddingLayer:
    def __init__(self,n_items,n_factors):
        self.n_items=n_items
        self.n_factors=n_factors
        
    def __call__(self,x):
        x=Embedding(self.n_items,self.n_factors,embeddings_initializer='he_normal',
                   embeddings_regularizer=l2(1e-6))(x)
        x=Reshape((self.n_factors,))(x)
        return x
    
def RecommenderV2(n_users,n_beers,n_factors,min_rating,max_rating):
    user=Input(shape=(1,))
    u=EmbeddingLayer(n_users,n_factors)(user)
    ub=EmbeddingLayer(n_users,1)(user)
    
    beer=Input(shape=(1,))
    b=EmbeddingLayer(n_beers,n_factors)(beer)
    bb=EmbeddingLayer(n_beers,1)(beer)
    
    x=Dot(axes=1)([u,b])
    x=Add()([x,ub,bb])
    x=Activation('sigmoid')(x)
    x=Lambda(lambda x: x*(max_rating-min_rating)+min_rating)(x)
    
    model=Model(inputs=[user,beer],outputs=x)
    opt=Adam(lr=0.001)
    model.compile(loss='mean_squared_error',optimizer=opt)
    
    return model
```

```python
model = RecommenderV2(n_users, n_beers, n_factors, min_rating, max_rating)
model.summary()
```

```python
history = model.fit(x=X_train_array, y=y_train, batch_size=64, epochs=5,
                    verbose=1, validation_data=(X_test_array, y_test))
```

```python
rate=pd.DataFrame(X_test_array).T
rate.columns=['userid','beerid']
rate=rate.reset_index()
rate=pd.merge(rate,pd.DataFrame(model.predict(X_test_array,batch_size=64)).reset_index(),on='index')
rate
```
